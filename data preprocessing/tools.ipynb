# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')

# : matrix for all the rows and : -1 for feature variables except last column. (exlcude upper bound)

X = dataset.iloc[:, :-1].values

# -1 for only last dependent variable

Y = dataset.iloc[:, -1].values
```

```
print(X)
```

```
print(Y)
```

## Taking care of missing data

```
# scikit-learn
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# 2nd and 3rd col will get selected
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```
print(X)
```

## Encoding categorical data

### Encoding the feature (independent) Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# column transform -> kind of transform
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder="passthrough")

# fit is func that will fit in eq or model to data transform() is a func rthat will apply the fit to a data set
# fit method is used to train the model with the data in simpler terms when ur feeding the data fit.transform is when ur manipulating the data

# fit transfer doesnt return the output as np array and for machine learning model its strict to output the result in np array so we convert it
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# we dont need to have an np array cuz this is dependent variable array it doesnt need to be non binary
Y = le.fit_transform(Y)
```

```
print(Y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split

# the x_train will correspond to the data to the y_train similarly with the x_test and y_test
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(Y_train)
```

```
print(Y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

# fit and transform will be applied to train set
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])

# transform only will be applied cuz we dont need new scalar of test set
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
